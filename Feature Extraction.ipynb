```python
#24306-34320 (-34321)
for x in range(24306,34321):
    
    
    img = io.imread('Intelligent Systems in Medicine/Project/ISIC2018_Task3_Training_Input/ISIC_00' + str(x) +'.jpg')

    # add average color information to dataframe
    average = img.mean(axis=0).mean(axis=0)
    
    df.loc[df['image_id'] == 'ISIC_00' + str(x), ['average_red']] = average[0]
    df.loc[df['image_id'] == 'ISIC_00' + str(x), ['average_green']] = average[1]
    df.loc[df['image_id'] == 'ISIC_00' + str(x), ['average_blue']] = average[2]

    
    #add Hu's moments information to dataframe
    
    moments = measure.moments_hu(color.rgb2grey(img))
    df.loc[df['image_id'] == 'ISIC_00' + str(x), ['moment_1']] = moments[0]
    df.loc[df['image_id'] == 'ISIC_00' + str(x), ['moment_2']] = moments[1]
    df.loc[df['image_id'] == 'ISIC_00' + str(x), ['moment_3']] = moments[2]
    df.loc[df['image_id'] == 'ISIC_00' + str(x), ['moment_4']] = moments[3]
    df.loc[df['image_id'] == 'ISIC_00' + str(x), ['moment_5']] = moments[4]
    df.loc[df['image_id'] == 'ISIC_00' + str(x), ['moment_6']] = moments[5]
    df.loc[df['image_id'] == 'ISIC_00' + str(x), ['moment_7']] = moments[6]


    # add texture information to dataframe
    gCoMat = feature.greycomatrix(img_as_ubyte(color.rgb2grey(img)),[2],[0], 256, symmetric = True, normed = True)

    df.loc[df['image_id'] == 'ISIC_00' + str(x), ['contrast']] = feature.greycoprops(gCoMat, prop='contrast')
    df.loc[df['image_id'] == 'ISIC_00' + str(x), ['dissimilarity']] = feature.greycoprops(gCoMat, prop='dissimilarity')
    df.loc[df['image_id'] == 'ISIC_00' + str(x), ['homogeneity']] = feature.greycoprops(gCoMat, prop='homogeneity')
    df.loc[df['image_id'] == 'ISIC_00' + str(x), ['energy']] = feature.greycoprops(gCoMat, prop='energy')
    df.loc[df['image_id'] == 'ISIC_00' + str(x), ['correlation']] = feature.greycoprops(gCoMat, prop='correlation')
    
    
    
    
```
